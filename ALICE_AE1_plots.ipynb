```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Anomalies_1 = np.loadtxt('Anomalies_d1')
Non_anomalies_1 = np.loadtxt('Non_anomalies_d1')    
Predictions = np.loadtxt('Predictions_d1')
dataset = np.loadtxt('dataset_d1')
History_train = np.loadtxt('History_train_d1')
History_validate = np.loadtxt('History_validate_d1')
```

```python
signal_length = 160

plt.figure(figsize=(20,4))
plt.plot(Anomalies_1, 'r')
plt.plot(Non_anomalies_1, 'b')
plt.plot(dataset, 'k')
plt.ylim([0,1])
plt.show()
```

```python
plt.figure(figsize=(20,4))
plt.plot(dataset, 'k')
plt.show()
```

```python
plt.figure(figsize=(20,7))
plt.plot(History_train, 'b', label='train')
plt.plot(History_validate, 'r', label='test')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(loc='best')
plt.show()
```
