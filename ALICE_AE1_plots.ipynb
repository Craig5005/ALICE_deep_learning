```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Anomalies_1 = np.loadtxt('Anomalies_d1a')
Non_anomalies_1 = np.loadtxt('Non_anomalies_d1a')    
Predictions = np.loadtxt('Predictions_d1a')
dataset = np.loadtxt('dataset_d1a')
History_train = np.loadtxt('History_train_d1a')
History_validate = np.loadtxt('History_validate_d1a')
```

```python
plt.figure(figsize=(20,4))
plt.plot(dataset, 'k')
plt.show()
```

```python
signal_length = 160

plt.figure(figsize=(20,4))
plt.plot(Anomalies_1, 'r')
plt.plot(Non_anomalies_1, 'b')
#plt.plot(dataset, 'k')
plt.ylim([0,1])
plt.show()
```

```python
buffers1 = np.zeros(signal_length)
buffers1[:] = np.nan

plt.figure(figsize=(20,4))
plt.plot(np.append(buffers1,Anomalies_1), 'r')
plt.plot(np.append(buffers1,Non_anomalies_1), 'b')
plt.plot(dataset, 'k')
plt.ylim([0,1])
plt.show()
```

```python
plt.figure(figsize=(20,7))
plt.plot(History_train, 'b', label='train')
plt.plot(History_validate, 'r', label='test')
plt.plot(np.gradient(History_validate), 'k')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(loc='best')
plt.show()
```

```python
def running_mean(x, N):
    cumsum = np.cumsum(np.insert(x, 0, 0)) 
    return (cumsum[N:] - cumsum[:-N]) / float(N)
results = running_mean(History_validate,3)

buffer1 = np.zeros(3)
buffer1[:] = np.nan

plt.figure(figsize=(20,7))
#plt.plot(results)
plt.plot(np.append(buffer1,np.full(len(results),0.25*np.std(np.gradient(results)))), 'k--')
plt.plot(np.append(buffer1,np.full(len(results),-0.25*np.std(np.gradient(results)))), 'k--')
plt.plot(np.append(buffer1,np.zeros(len(results))), 'k', linewidth=3)
plt.plot(np.append(buffer1,np.gradient(results)), 'r')
plt.show()
```

```python
if np.gradient(results)[-1] > np.std(np.gradient(results)) or np.gradient(results)[-1] < -np.std(np.gradient(results)):
    print 'Anomaly'
else:
    print 'Non-anomaly'
```
